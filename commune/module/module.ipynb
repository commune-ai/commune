# Module Management

```python
import commune as c
```

## Finding Your Module

```python
# list all of the modules
c.modules()[:10]
```

```python
# to search for a specific module, use the search function
for k in ['model.llama', 'data', 'demo', 'hf']:
    c.print('Searching for ', k)
    c.print( c.modules(k))
```

## Module Management

```python
demo = c.module('demo')

c.print('## Code for demo module')
c.print(demo.code())
```

```python
# see the config
demo.config()
```

```python
# list the functions

demo.fns()
```

```python
# search for a function
demo.fns('bro')
```

```python
# get the function schema
demo.schema('bro')
```

## Serving

```python
# serve the module
demo.serve(tag='tag1') # tag is optional, but useful for versioning
```

```python
c.servers()
```

```python
# get the logs of your served module
c.print(c.logs('demo::tag1', mode='local'))
```

```python
demo_client = c.connect('demo::tag1')
demo_client.info()
```

```python
c.restart('demo::tag1') # restart the module
c.kill('demo::tag1') # kill the module
```
